```python
#checking if a number is a power of another
```

```python
def ispower(x, y):
    if x == 1:
        return y == 1
    pow = 1
    while pow < y:
        pow *= x
    return pow == y

print("True" if ispower(4,64) else "False")
print("True" if ispower(10,1) else "False")
print("True" if ispower(1,20) else "False")
print("True" if ispower(2,128) else "False")
print("True" if ispower(2,30) else "False")
```
