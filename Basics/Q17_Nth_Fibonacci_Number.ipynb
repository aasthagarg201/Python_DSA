```python
#Nth Fibonacci Number
#Fibonacci series is a sequence of numbers in which each number is the sum of two previous numbers.
#1. First number = 0 
#2. Second number = 1
# Every next number = Previous no. + no. before it
```

```python
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n - 1) + fibonacci(n - 2)
n=6
print(fibonacci(n))
```
