```python
def divisor(num):
    cnt = 0

    for i in range(1, num + 1):
        if num % i == 0:
            cnt += 1
            
    return cnt
    
def threedivisors(queries):
    ans = []
    
    for n in queries:
        total = 0
        for num in range(1, n +1):
            if divisor(num) == 3:
                total += 1
                
        ans.append(total)
        
    return ans
queries = [6, 10, 100]
ans = threedivisors(queries)
print(ans)
```
