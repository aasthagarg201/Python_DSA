```python
def reversedigits(n):
    x = str(n)
    x = list(x)
    x.reverse()
    x = ''.join(x)

    n = int(x)
    return n

num = 4562
print(reversedigits(num))
```

```python
n=1024
xyz = 0
while (n > 0):
    a = n % 10
    xyz = xyz * 10 + a
    n = n // 10

print(xyz)
```
