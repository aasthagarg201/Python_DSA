```python
def addfraction(a, b):
    den = gcd(a[1], b[1])
    
    den = (a[1] * b[1]) // den

    num = a[0] * (den // a[1]) + b[0] * (den // b[1])

    common_factor = gcd(num, den)
    num //= common_factor
    den //= common_factor
    return [num, den]

a = [1, 2]
b = [3, 2]
ans = addfraction(a, b)
print(f'{ans[0]}, {ans[1]}')
```
