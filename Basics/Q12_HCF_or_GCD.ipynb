```python
#Finding HCF(highest common factor) or GCD(greatest common divisor)
```

```python
def hcf(a, b):
    if a == 0:
        return b
    if b == 0:
        return a
    if a == b:
        return a
    if a > b:
        return hcf(a-b, b)
    return hcf(a, b-a)

a = 36
b = 28
print(hcf(a, b))
```
