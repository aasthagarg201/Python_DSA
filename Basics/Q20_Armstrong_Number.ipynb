```python
#An armstrong number is a number that is equal to the sum of its own digits, where each digit is raised to the power of the total number of digits.
#Formula: For an n-digit number:
# Number=∑(each digit)^n     [where n is the number of digits]
```

```python
def armstrong(n):
    power = len(str(n))
    total = 0

    for digit in str(n):
        total += int(digit) ** power
        
    return total == n

n = 153
if armstrong(n):
    print("True")
else:
    print("False")
```
