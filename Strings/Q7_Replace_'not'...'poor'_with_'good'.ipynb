```python
#Replacing 'not'....'poor' with 'good'
```

```python
def not_poor(str1):
    x = str1.find('not')
    y = str1.find('poor')
    if y>x and x>0 and y>0:
        str1 = str1.replace(str1[x:(y+4)],'good')
        return str1
    else:
        return str1

print(not_poor('The lyrics is not that poor!'))
print(not_poor('The lyrics is poor!'))
```
